### №3

```python
plt.figure(figsize=(10, 10))
for i in range(3):
    plt.subplot(2, 3, i + 1)
    img = np.copy(duck)
    for j in range(3):
        if i != j:
            img[:, :, j] = 0
    io.imshow(img)
for i in range(3):
    plt.subplot(2, 3, i + 4)
    img = np.copy(duck)
    img[:, :, i] = 0
    io.imshow(img)
```

Первая строка - отдельные RGB каналы, вторая - получаемые из исходного занулением одного из каналов

### №4

```python
avg = np.mean(duck[:, :, :3], axis=2) * 255
avg = avg.astype('uint8')
print (avg.shape)
io.imshow(avg)
```

```python
img_gray = skimage.color.rgb2gray(duck) * 255
img_gray = img_gray.astype('uint8')
io.imshow(img_gray)
```

```python
img_conc = np.concatenate([avg, img_gray], axis=1)
io.imshow(img_conc)
```

Как мы видим из сконкатенированного изображения, перевод в ЧБ получается более четким при использовании skimage.color.rgb2gray (правое изображение), нежели при использовании простого усреднения (левое изображение)
